```python
import os
import pandas as pd
import pickle
from datetime import datetime
```

```python
all_files = os.listdir('saved_superpoints/pickle/test_realdata')
```

```python
all_abs_path = []
for path, folders, files in os.walk('saved_superpoints/pickle/test_realdata'):
    for file in files:
        all_abs_path.append(os.path.abspath(os.path.join(path,file)).replace(os.sep,'/'))
    # print(path, folder, file, '\n')
```

```python
df = pd.DataFrame({'file_name':all_files, 'abspath':all_abs_path})
df
```

```python
df['location'] = range(1, len(df)+1)
```

```python
df['date_taken'] = datetime.now().strftime('%m/%d/%Y')
```

```python
df
```

```python
df.to_csv('superpoints_database.csv', index=True)
```
